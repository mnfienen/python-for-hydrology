# 07a: Using functions to solve an equation-- The Theis example
In this exercise we will implement the Theis equation in Python using functions, to evaluate drawdown in hydraulic head from pumping at a well.


### Background
The Theis (1935) equation is used to calculate drawdown for two-dimensional radial groundwater flow to a point source in an infinite, homogeneous aquifer. The Theis equation was derived from heat transfer literature (with the mathematical help of C.I. Lubin) and is defined as:

\begin{equation}
s = \frac{Q}{4 \pi T} W(u)
\end{equation}

where  
$s$ is drawdown [L],  
$Q$ is the pumping rate [L$^3$/T],  
$T$ is the aquifer transmissivity [L$^2$/T],  
$u$ is a dimensionless time parameter [unitless], and  
$W(u)$ is the Well function (exponential integral $E_1$) [unitless]. The exponential integral is available in ``scipy.special`` as the ``exp1()`` function.

The dimensionless time parameter is defined as:

\begin{equation}
u = \frac{r^2S}{4Tt}
\end{equation}

where  
$r$ is the distance from the pumping well to a point where drawdown is observed [L],   
$S$ is storativity [unitless], and  
$t$ is the time since pumping began. 

Storativity is defined as:

\begin{equation}
S = S_s b
\end{equation}

where  
$S_s$ is specific storage [1/L] and   
$b$ is the thickness of the aquifer.

### AI Approach

**Prompt used:**

> *"Write a Python function called `theis_drawdown` that computes drawdown using the Theis equation. The function should accept: pumping rate Q (m³/day), transmissivity T (m²/day), storativity S (unitless), distance from pumping well r (m), and time since pumping began t (days). Use scipy.special.exp1 for the well function W(u). Include a numpy-style docstring. Then compute drawdown at r=1000m, t=10 days with Q=4088 m³/day, T=1000 m²/day, S=3e-4."*

**AI-generated code:**

```python
# AI-GENERATED CODE (pre-captured)
import numpy as np
from scipy.special import exp1

def theis_drawdown(Q, T, S, r, t):
    """Compute drawdown using the Theis (1935) equation.

    Parameters
    ----------
    Q : float
        Pumping rate in m³/day.
    T : float
        Aquifer transmissivity in m²/day.
    S : float
        Storativity (dimensionless).
    r : float or numpy.ndarray
        Distance from the pumping well in meters.
    t : float or numpy.ndarray
        Time since pumping began in days.

    Returns
    -------
    float or numpy.ndarray
        Drawdown in meters.
    """
    u = (r**2 * S) / (4.0 * T * t)
    s = (Q / (4.0 * np.pi * T)) * exp1(u)
    return s
```

### What Could Go Wrong

AI assistants can make subtle errors with the Theis equation. Two common failure modes:

**1. Wrong exponential integral:** Some AI implementations use `exp(-u)/u` as an approximation instead of the proper `scipy.special.exp1(u)`. This gives reasonable results for small $u$ but diverges for larger values.

**2. Sign error in the well function:** The exponential integral $E_1(u)$ is sometimes confused with the related function $-Ei(-u)$. While mathematically equivalent for positive $u$, using the wrong function or adding a stray negative sign will flip the drawdown to negative (water levels going *up* from pumping — physically impossible).

```python
# Demonstration: what happens with a sign error
from scipy.special import expi

u = (r**2 * S) / (4.0 * T * t)

# Correct: using exp1(u)
W_correct = exp1(u)

# Wrong: accidentally using expi(u) without the sign correction
W_wrong = expi(u)  # This is Ei(u), not E1(u)!

s_correct = (Q / (4.0 * np.pi * T)) * W_correct
s_wrong = (Q / (4.0 * np.pi * T)) * W_wrong

print(f"W(u) correct [exp1]:  {W_correct:.6f}")
print(f"W(u) wrong [expi]:    {W_wrong:.6f}")
print(f"")
print(f"Drawdown correct: {s_correct:.6f} m")
print(f"Drawdown wrong:   {s_wrong:.6f} m")
print(f"")
print("-> The wrong function gives a completely different value.")
print("   Always verify against a known solution!")
```

### 🤖 Debugging Exercise: Transmissivity vs. Hydraulic Conductivity

The following function was generated by an AI assistant in response to this prompt:

> *"Write a Python function to compute Theis drawdown. The aquifer has hydraulic conductivity K=10 m/day, thickness b=100 m, storativity S=3e-4. Compute drawdown at r=1000 m, t=10 days for Q=4088 m³/day."*

The AI produced code that runs without errors — but it has a **subtle parameter confusion bug**. Your job is to find it.

**Ground truth:** With $K = 10$ m/day and $b = 100$ m, the transmissivity is $T = K \times b = 1000$ m²/day. The correct drawdown is **1.40636669 m**.

```python
# AI-GENERATED CODE (pre-captured) — contains a bug!
import numpy as np
from scipy.special import exp1

def theis_drawdown_buggy(Q, K, b, S, r, t):
    """Compute drawdown using the Theis equation.

    Parameters
    ----------
    Q : float
        Pumping rate in m³/day.
    K : float
        Hydraulic conductivity in m/day.
    b : float
        Aquifer thickness in meters.
    S : float
        Storativity (dimensionless).
    r : float
        Distance from pumping well in meters.
    t : float
        Time since pumping began in days.

    Returns
    -------
    float
        Drawdown in meters.
    """
    # BUG: uses K directly instead of computing T = K * b
    u = (r**2 * S) / (4.0 * K * t)
    s = (Q / (4.0 * np.pi * K)) * exp1(u)
    return s
```

```python
# Test the buggy function
Q = 4088.0   # m³/day
K = 10.0     # m/day (hydraulic conductivity)
b = 100.0    # m (aquifer thickness)
S = 3e-4     # unitless
r = 1000.0   # m
t = 10.0     # days

s_buggy = theis_drawdown_buggy(Q, K, b, S, r, t)
T_correct = K * b  # T = K * b = 1000 m²/day
s_expected = 1.40636669

print(f"AI function result:  {s_buggy:.8f} m")
print(f"Expected (correct):  {s_expected:.8f} m")
print(f"")
print(f"The AI used K = {K} m/day directly in the Theis equation,")
print(f"but the equation requires T (transmissivity) = K × b = {T_correct} m²/day.")
print(f"")
print(f"The result is off by a large factor because the function")
print(f"never multiplied K by the aquifer thickness b.")
print(f"")
print(f"Can you find where in the code K should be replaced with T = K * b?")
```

**Why this matters:** Confusing transmissivity ($T$, units of m²/day) with hydraulic conductivity ($K$, units of m/day) is one of the most common errors in groundwater modeling. The Theis equation requires $T$, which equals $K \times b$ (hydraulic conductivity times aquifer thickness). When an AI sees both $K$ and $b$ in the prompt but uses $K$ directly in the equation, the drawdown will be wrong by a factor related to the aquifer thickness. In this case, using $K = 10$ instead of $T = 1000$ produces a drawdown that is far too large. Always check that the correct aquifer property is used in each equation — the AI won't catch this dimensional inconsistency for you.
