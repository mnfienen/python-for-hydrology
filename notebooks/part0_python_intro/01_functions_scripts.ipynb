## Why use functions?

functions are a great coding tool! They allow you to set up reusable bits of code that you might need over and over. Using functions, you only have to write the code once, and if you need to change it, only change it once.  

We can create a function that writes the Fibonacci series to an arbitrary boundary:

````python
def fib(n):
    """Print a Fibonacci series up to n."""
    a, b = 0, 1
    while a < n:
        print(a)
        a, b = b, a + b
````

````python
a, b = 0, 1
````

````python
help(fib)
````

Now call the function we just defined:

````python
fib(2000)
````

The keyword ```def``` introduces a function definition. It must be followed by the function name and the parenthesized list of formal parameters or arguments. The statements that form the body of the function start at the next line, and must be indented.


The first statement of the function body can optionally be a string literal; this string literal is the function's documentation string, or docstring. There are tools which use docstrings to automatically produce online or printed documentation, or to let the user interactively browse through code. It's good practice to include docstrings in code that you write, so make a habit of it.


The variables that are formed and used within a function can only be used within the function, unless they are ```return```-ed.


**More detailed information**: *the execution of a function introduces a new symbol table used for the local variables of the function.  all variable assignments in a function store the value in the local symbol table; whereas variable references first look in the local symbol table, then in the local symbol tables of enclosing functions, then in the global symbol table, and finally in the table of built-in names. Thus, global variables cannot be directly assigned a value within a function (unless named in a global statement), although they may be referenced.*

*The actual parameters (arguments) to a function call are introduced in the local symbol table of the called function when it is called; thus, arguments are passed using call by value (where the value is always an object reference, not the value of the object). Additionally, when a function calls another function, a new local symbol table is created for that call.*

A function definition introduces the function name, which has a ```type``` recognized by the interpreter as a user-defined function. 

This value can be assigned to another name which can then also be used as a function. This serves as a general renaming mechanism.

It is simple to write a function that returns a list of the numbers of the Fibonacci series, instead of printing it:

````python
def fib2(n):  # return Fibonacci series up to n
    """Return a list containing the Fibonacci series up to n."""
    result = []
    a, b = 0, 1
    while a < n:
        result.append(a)  # see below
        a, b = b, a + b
    return result
````

````python
fib2(100)
````

````python
f100 = fib2(100)  # call it
f100  # write the result
````

This example demonstrates some new Python features:

The ```return``` statement returns with a value from a function. ```return``` without an expression argument returns ```None```. Falling off the end of a function also returns ```None```.

## An object oriented preview aside: <span style="color:red">terminology alert!!!</span>

The statement ```result.append(a)``` calls a method of the ```list```-type object instance ```result```. 

**A method is simply a function that "belongs" to an object** and is named obj.methodname, where obj is some object (this may be an expression), and methodname is the name of a method that is defined by the object’s type. 

Different object types will have different methods associated with them. Methods of different types may have the same name without causing ambiguity. (It is possible to define your own object types and methods using ```Classes```).

The method ```append()``` shown above is defined for ```list``` objects; it adds a new element at the end of the list. In this example it is equivalent to ```result = result + [a]```, but more efficient.

## back to functions

It is also possible to define functions with a variable number of arguments. In Python, there are **three forms of function arguments** which can be combined: **default, keyword, or positional**. Each form is used frequently.

The most user-friendly function argument form is to **specify an argument with a default value**. This creates a function that can be called with fewer arguments, but also allows greater flexility in controlling the behavior within the function itself. 

For example:

````python
import math


def thiem(T, r, Q=1000.0, R=1.0e10, h0=0.0):
    """
    A very simple example function
    with a mixture of argument types.
    Solves the Thiem equation:

    h = (Q/2piT)*(ln(R/r)) + h0

    Parameters
    ----------

    T: transmissivity
    r: distance from pumping to observation
    Q: pumping rate
    R: distance to "zero" influence
    h0: initial head

    Returns
    -------
    h: head
    """
    first_term = Q / (2.0 * 3.14159 * T)
    second_term = math.log(R / r)
    return (first_term * second_term) + h0
````

````python
help(thiem)
````

This function *requires* both ```T``` and ```r```.  All the rest use the defaults if other values are not explicitly passed in. The default values are evaluated at the point of function definition in the defining scope.

The mandatory arguments `T` and `r` can be defined based on their **position** in the function definition:

````python
#  T=100.0, r=300.0
thiem(100.0, 300.0)
````

They can also be defined **explicitly by their keyword**, and thus be passed in any order.

````python
#  T=100.0, r=300.0
thiem(r=300.0, T=100.0)
````

Passing optional argument `Q` using it's implied position in the function definition: `T`=100.0, `r`=300.0, `Q`=500.0

````python
thiem(100.0, 300.0, 500)
````

Giving one of the optional arguments using the explicit argument name: `T`=100.0, `r`=300.0, `Q`=500.0

````python
thiem(100.0, 300.0, Q=500.0)
````

## Scripts

If you quit from the Python interpreter and enter it again, the definitions you have made (functions and variables - _e.g._ ALL YOUR HARD WORK) are lost. Therefore, if you want to write a somewhat longer program, you are better off using a text editor to prepare the input for the interpreter and running it with that file as input instead. This is known as creating a script. As your program gets longer, you may want to split it into several files for easier maintenance. You may also want to use a handy function that you’ve written in several programs without copying its definition into each program.

To support this, Python has a way to put definitions in a file and use them in a script or in an interactive instance of the interpreter. Such a file is called a module; definitions from a module can be imported into other modules or into the main module (the collection of variables that you have access to in a script executed at the top level and in calculator mode).

A module is a file containing Python definitions and statements. The file name is the module name with the suffix .py appended. Within a module, the module’s name (as a string) is available as the value of the global variable __name__.

A module can contain executable statements as well as function definitions. These statements are intended to initialize the module. They are executed only the first time the module name is encountered in an import statement. (They are also run if the file is executed as a script.)

Each module has its own private symbol table, which is used as the global symbol table by all functions defined in the module. Thus, the author of a module can use global variables in the module without worrying about accidental clashes with a user's global variables. On the other hand, if you know what you are doing you can touch a module's global variables with the same notation used to refer to its functions, modname.itemname.

Modules can import other modules. It is customary but not required to place all import statements at the beginning of a module (or script, for that matter). The imported module names are placed in the importing module’s global symbol table.

## 🤖 AI Sidebar: Writing Functions with AI Assistance

*This sidebar demonstrates how an AI coding assistant approaches the same kind of task you just practiced — writing Python functions. All AI-generated code below was captured ahead of time; no live AI calls are made.*

### The Task

Earlier in this notebook, you wrote and used functions like `thiem()` and `thiem_list()` that take parameters, perform calculations, and return results. You practiced using positional arguments, keyword arguments, default values, and docstrings.

Now let's see how an AI assistant handles a similar task: **writing a unit conversion function** for hydrology work. Unit conversions are a common source of errors in scientific code, so this is a good test of whether AI-generated code can be trusted.

### AI Approach

**Prompt used:**

> *"Write a Python function called `convert_flow` that converts streamflow between common hydrology units. It should accept a flow value, a `from_unit` string, and a `to_unit` string. Support these units: cubic feet per second (cfs), cubic meters per second (cms), gallons per minute (gpm), and liters per second (lps). Include a docstring with parameter descriptions. Return the converted value."*

**AI-generated code:**

````python
# AI-GENERATED CODE (pre-captured)
def convert_flow(flow, from_unit, to_unit):
    """Convert streamflow between common hydrology units.

    Parameters
    ----------
    flow : float
        The flow value to convert.
    from_unit : str
        Source unit. One of: "cfs", "cms", "gpm", "lps"
    to_unit : str
        Target unit. One of: "cfs", "cms", "gpm", "lps"

    Returns
    -------
    float
        The converted flow value.
    """
    # Conversion factors to cubic meters per second (cms)
    to_cms = {
        "cfs": 0.0283168,
        "cms": 1.0,
        "gpm": 6.309e-5,
        "lps": 0.001,
    }

    # Convert to cms first, then to target unit
    flow_cms = flow * to_cms[from_unit]
    return flow_cms / to_cms[to_unit]
````

Let's test the AI-generated function against known conversion values:

````python
# Test against known values
# 1 cfs = 0.0283168 cms (this is a well-known conversion factor)
result_1 = convert_flow(1.0, "cfs", "cms")
expected_1 = 0.0283168
print(f"1 cfs -> cms: {result_1:.7f}  (expected: {expected_1:.7f})")

# 100 cfs = 2831.68 lps
result_2 = convert_flow(100.0, "cfs", "lps")
expected_2 = 28.3168
print(f"100 cfs -> lps: {result_2:.4f}  (expected: {expected_2:.4f})")

# Round-trip: convert 500 gpm -> cms -> gpm should return 500
result_3 = convert_flow(convert_flow(500.0, "gpm", "cms"), "cms", "gpm")
print(f"500 gpm -> cms -> gpm: {result_3:.4f}  (expected: 500.0000)")

# Identity: converting to the same unit should return the same value
result_4 = convert_flow(42.0, "cms", "cms")
print(f"42 cms -> cms: {result_4:.4f}  (expected: 42.0000)")
````

### What Could Go Wrong

The AI-generated `convert_flow` function looks clean and works for valid inputs. But watch what happens with edge cases:

````python
# Problem 1: No input validation — negative flow is physically meaningless
print("Negative flow:", convert_flow(-100.0, "cfs", "cms"))
print("  -> No error! Negative streamflow is physically impossible.")
print()

# Problem 2: Unknown unit gives a confusing KeyError, not a helpful message
try:
    convert_flow(100.0, "m3/s", "cfs")  # common alternative name for cms
except KeyError as e:
    print(f"Unknown unit error: KeyError({e})")
    print("  -> The AI didn't handle unrecognized unit strings gracefully.")
print()

# Problem 3: No type checking — passing a string silently breaks
try:
    convert_flow("one hundred", "cfs", "cms")
except TypeError as e:
    print(f"Type error: {e}")
    print("  -> The AI didn't validate that flow is a number.")
````

**Key takeaway:** AI-generated functions typically handle the "happy path" well but skip **edge case handling** and **input validation**. In scientific code, these gaps can lead to silent errors — the code runs without crashing but produces physically meaningless results.

A more robust version would validate inputs:

````python
# Improved version with input validation (what a domain expert would add)
def convert_flow_safe(flow, from_unit, to_unit):
    """Convert streamflow between common hydrology units.

    Parameters
    ----------
    flow : float
        The flow value to convert. Must be non-negative.
    from_unit : str
        Source unit. One of: "cfs", "cms", "gpm", "lps"
    to_unit : str
        Target unit. One of: "cfs", "cms", "gpm", "lps"

    Returns
    -------
    float
        The converted flow value.

    Raises
    ------
    ValueError
        If flow is negative or units are not recognized.
    """
    to_cms = {
        "cfs": 0.0283168,
        "cms": 1.0,
        "gpm": 6.309e-5,
        "lps": 0.001,
    }

    if not isinstance(flow, (int, float)):
        raise ValueError(f"flow must be a number, got {type(flow).__name__}")
    if flow < 0:
        raise ValueError(f"flow must be non-negative, got {flow}")
    if from_unit not in to_cms:
        raise ValueError(
            f"Unknown from_unit "{from_unit}". "
            f"Supported units: {list(to_cms.keys())}"
        )
    if to_unit not in to_cms:
        raise ValueError(
            f"Unknown to_unit "{to_unit}". "
            f"Supported units: {list(to_cms.keys())}"
        )

    flow_cms = flow * to_cms[from_unit]
    return flow_cms / to_cms[to_unit]


# Now the error messages are clear and helpful
try:
    convert_flow_safe(-100.0, "cfs", "cms")
except ValueError as e:
    print(f"Caught: {e}")

try:
    convert_flow_safe(100.0, "m3/s", "cfs")
except ValueError as e:
    print(f"Caught: {e}")
````

### When to Use AI Here

- ✅ **Well-suited for AI:** Generating function boilerplate and docstrings, looking up standard conversion factors, creating the dictionary-based conversion pattern
- ✅ **Well-suited for AI:** Writing repetitive test cases to verify the function
- ⚠️ **Requires manual understanding:** Adding input validation for domain-specific constraints (e.g., flow must be non-negative)
- ⚠️ **Requires manual understanding:** Verifying that conversion factors are correct for *your* specific use case (e.g., are you converting instantaneous or mean daily flow?)
- ⚠️ **Requires manual understanding:** Deciding which units to support based on your project needs
